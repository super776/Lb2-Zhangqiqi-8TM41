```python
import pandas as pd

# 读取原始数据
df = pd.read_csv("../data/train_(A+B).csv")

# 提取 Type==1 的数据，保留原标记
df_type_1 = df[df['Type'] == 1]

# 提取其它类型数据，并统一标记为 Type=0
df_other = df[df['Type'] != 1].copy()
df_other['Type'] = 0

# 将两个部分上下拼接，并按顺序排列（Type=1在前）
df_sorted = pd.concat([df_type_1, df_other], ignore_index=True)

# 保存为新文件
df_sorted.to_csv("train_binary_sorted.csv", index=False)

# 显示前几行检查
df_sorted.head()
```
